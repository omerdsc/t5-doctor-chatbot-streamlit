```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import nltk
import re

# ðŸ”½ Gerekli NLTK modÃ¼lleri
nltk.download("punkt")
nltk.download("stopwords")
from nltk.corpus import stopwords

# ðŸ“¥ Veri setini oku
df = pd.read_csv("train_google.csv")

# ðŸ‘€ Ä°lk birkaÃ§ veriye gÃ¶z at
print(df.head())

# ðŸ§¼ BoÅŸ kayÄ±tlarÄ± temizle ve tipleri dÃ¼zelt
df = df.dropna(subset=["Question", "Answer"])
df["Question"] = df["Question"].astype(str).str.strip()
df["Answer"] = df["Answer"].astype(str).str.strip()

# ðŸ§¹ TÃ¼rkÃ§e stopwords listesi
stop_words = set(stopwords.words("turkish"))

# ðŸ§  Temizleme fonksiyonu (stopwords dahil)
def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\@\w+|\#", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)

# âœ… Temiz metinleri yeni sÃ¼tunlara yaz
df["clean_question"] = df["Question"].apply(clean_text)
df["clean_answer"] = df["Answer"].apply(clean_text)
```

```python
df["question_length"] = df["clean_question"].apply(lambda x: len(x.split()))
df["answer_length"] = df["clean_answer"].apply(lambda x: len(x.split()))

# ðŸŽ¨ Soru Uzunluk GrafiÄŸi
sns.histplot(df["question_length"], bins=30, kde=True)
plt.title("Soru UzunluÄŸu DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Kelime SayÄ±sÄ±")
plt.ylabel("SÄ±klÄ±k")
plt.show()

# ðŸŽ¨ Cevap Uzunluk GrafiÄŸi
sns.histplot(df["answer_length"], bins=30, kde=True, color="orange")
plt.title("Cevap UzunluÄŸu DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Kelime SayÄ±sÄ±")
plt.ylabel("SÄ±klÄ±k")
plt.show()
```

```python
# Soru Wordcloud
text = " ".join(df["clean_question"].tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Sorularda GeÃ§en En SÄ±k Kelimeler")
plt.show()

# Cevap Wordcloud
text = " ".join(df["clean_answer"].tolist())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Cevaplarda GeÃ§en En SÄ±k Kelimeler")
plt.show()
```

```python
from datasets import Dataset

# Dataset objesine Ã§evir
hf_dataset = Dataset.from_pandas(df[["clean_question", "clean_answer"]].rename(
    columns={"clean_question": "input_text", "clean_answer": "target_text"}))
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_ckpt = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt)
```

```python
def preprocess_function(examples):
    inputs = tokenizer(examples["input_text"], truncation=True, padding="max_length", max_length=256)
    targets = tokenizer(examples["target_text"], truncation=True, padding="max_length", max_length=256)

    inputs["labels"] = targets["input_ids"]
    return inputs

tokenized_dataset = hf_dataset.map(preprocess_function, batched=True)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./doktorbot_model",
    evaluation_strategy="no",
    per_device_train_batch_size=8,
    num_train_epochs=3,
    logging_steps=100,
    save_steps=500,
    save_total_limit=1,
    fp16=True,
    report_to="none"
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer,
)

trainer.train()
```

```python
trainer.save_model("./doktorbot_model")
tokenizer.save_pretrained("./doktorbot_model")
```

```python
def generate_answer(question):
    inputs = tokenizer("Soru: " + question, return_tensors="pt", truncation=True, padding=True)
    output = model.generate(**inputs, max_new_tokens=128)
    return tokenizer.decode(output[0], skip_special_tokens=True)

generate_answer("Burun spreyleri poliplere iyi gelir mi?")
```
